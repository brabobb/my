```python
from selenium import webdriver as wd
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
#셀레늄 드라이버로 url 가져오기
driver = wd.Chrome(executable_path = 'chromedriver.exe')
url = 'https://www.youtube.com/channel/UCfpaSruWW3S4dibonKXENjA/videos'
driver.get(url)
```

```python
#동영상 있는 body 부분
body_tag = driver.find_element_by_tag_name('body')
```

```python
# 모든 동영상 스크롤 내려서 가져오기
while True:
    last = driver.execute_script('return document.documentElement.scrollHeight')
    print('현재 화면 크기:', last)

    for i in range(10):
        body_tag.send_keys(Keys.END)
        time.sleep(3)
    new = driver.execute_script('return document.documentElement.scrollHeight')
    print('스크롤 후 화면 크기:', new)
    print('-' * 60)
    
    if last == new:
        break
```

```python
#페이지 소스 가져오기
source = driver.page_source
parser = bs(source,'html.parser')
parser.text
```

```python
#업로드된 영상의 모든 정보 가져오기
videos = parser.find_all(id = 'dismissible')
videos
```

```python
#영상 제목
titles = []

for one_v in videos:
    title = one_v.find(id = 'video-title')
    if len(title.text.strip()) > 0:
        titles.append(title.text.strip())
print(titles)
print(len(titles))
```

```python
#영상 시간(분)
time_lst = []
for video in videos:
    run_time = video.find('span', 
                          {'class':'style-scope ytd-thumbnail-overlay-time-status-renderer'})
    time_lst.append(run_time.text.strip())

run_time = []
for time in time_lst:
    run_time.append(int(time.split(':')[0]))
print(run_time)
print(len(run_time))
```

```python
#조회수 
view_num_lst = []

import re

for video in videos:
    view_num = video.find('span', 
                          {'class':'style-scope ytd-grid-video-renderer'})
    print(view_num.text)
    view_num = view_num.text.split(' ')[1]
    view_num = view_num.split('회')[0]
    view_num_lst.append(view_num.strip())

for i in range(len(view_num_lst)):
    txt = view_num_lst[i]
    
    if '.' not in txt:
        txt = int(txt[0:2]) * 10000
        view_num_lst[i] = txt
    
    elif '.' in txt:
        txt = int(float(txt[0:3]) * 10000)
        view_num_lst[i] = txt

print(view_num_lst)
print(len(view_num_lst))
```

```python
data = {'title':titles, 'run_time':run_time, 'view_num':view_num_lst}
df = pd.DataFrame(data)
df
```

```python
import graphviz
from sklearn.tree import export_graphviz
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import warnings
```

```python
# DecisionTree Classifier 생성
dt_clf = DecisionTreeClassifier(random_state=10)

# 붓꽃 데이터 로딩, 학습과 테스트 데이터 세트로 분리
iris_data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target,
                                                   test_size=0.2, random_state=10)

#학습
dt_clf.fit(X_train, y_train)
```

```python
from sklearn.tree import export_graphviz

# export_graphviz() 호출 결과로 out_file로 지정된 tree.dot 파일을 생성
export_graphviz(dt_clf, out_file="tree.dot", class_names=iris_data.target_names, \
               feature_names = iris_data.feature_names, impurity=True, filled=True)
               

with open("tree.dot") as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```
