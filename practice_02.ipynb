```python
x = np.arange(-9, 10)
y = x ** 2

plt.plot(x, y, linestyle = "-", marker = "o")
plt.show()
```

```python
x = np.arange(-9, 10)
y1 = x ** 2
y2 = -x

plt.plot(x, y1, linestyle = "-.", marker = "*", color = "red", label = "y = x * x")
plt.plot(x, y2, linestyle = ":", marker = "o", color = "blue", label = "y = -x")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(
    shadow = True,
    borderpad = 1
)
plt.show()
```

```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)

result = []

for row in data:
    if row[-1] != '':
        if int(row[0].split('-')[0]) >= 1983:
            if int(row[0].split('-')[1]) == 2 and int(row[0].split('-')[2]) == 14:
                result.append(float(row[-1]))
                
plt.plot(result, 'lightpink')
plt.show()
```

```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)

aug = []
jan = []

for row in data:
    if row[-1] != '':
        
        if row[0].split('-')[1] == '08':
            aug.append(float(row[-1]))
        
        elif row[0].split('-')[1] == '01':
            jan.append(float(row[-1]))

plt.boxplot(aug)
plt.boxplot(jan)
plt.show()
```

```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)

aug = []
jan = []

for row in data:
    if row[-1] != '':
        
        if row[0].split('-')[1] == '08':
            aug.append(float(row[-1]))
        
        elif row[0].split('-')[1] == '01':
            jan.append(float(row[-1]))

plt.boxplot([aug, jan])
plt.show()
```
