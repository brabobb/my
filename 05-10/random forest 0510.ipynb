```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd
```

```python
#dict 형식
#data : feature 값 / target : target 값
load_iris()
```

```python
df = pd.DataFrame(
                  data = load_iris()['data'],
                  columns = ['sepal length', 'sepal width', 'petal length', 'petal width']
)
df
```

```python
#data frame에 컬럼 추가
df['target'] = load_iris()['target']
df
```

```python
X = df.loc[:, 'sepal length':'petal width']
X
```

```python
y = df.loc[:, 'target':]
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
#leaf노드에서 컬럼사용 비율을 계산
#많이 사용된 컬럼일수록 지니지수가 낮다는 뜻
rfc = RandomForestClassifier(oob_score = True)
```

```python
rfc.fit(X_train, y_train)
predict = rfc.predict(X_test)
```

```python
#서로 다른 decision tree가 100개
len(rfc.estimators_)
```

```python
rfc.score(X_test, y_test)
```

```python
from sklearn import tree
import pydotplus
from PIL import Image
from io import BytesIO
```

```python
rfc_dot_data = tree.export_graphviz(
                        rfc.estimators_[0], 
                        feature_names = X_train.columns,
                        class_names = ['Iris-Setosa', 'Iris-Versicolor', 'Iris-Virginica'],
                        filled = True, rounded = True, special_characters = True
)

rfc_graph = pydotplus.graph_from_dot_data(rfc_dot_data).create_png()
Image.open(BytesIO(rfc_graph))
```

```python
#100개 중 1번~10번을 pdf로 저장
for i in range(1, 11):
    rfc_dot_data = tree.export_graphviz(
                        rfc.estimators_[i], 
                        feature_names = X_train.columns,
                        class_names = ['Iris-Setosa', 'Iris-Versicolor', 'Iris-Virginica'],
                        filled = True, rounded = True, special_characters = True)

    pydotplus.graph_from_dot_data(rfc_dot_data).write_pdf('tree' + str(i) + '.pdf')
```
