```python
import csv
import matplotlib.pyplot as plt

#원하는 지역 입력 받기
addr = input('찾고 싶은 지역의 이름을 알려주세요: ')

#파일 오픈
f = open('gender.csv')
data = csv.reader(f)

female = []
male = []

for row in data:
    if addr in row[0]:
        
        #여성 인구 분포도(오른쪽)
        for i in range(0, 101):
            female.append(int(row[-(i+1)]))
            
        #남성 인구 분포도(왼쪽)
        for i in range(0, 101):
            male.append(-int(row[i+3]))

#title에 바로 addr 넣으면 오류남
header = addr + ' 지역의 남녀 성별 인구 분포'     

#여성 인구분포도 역정렬
female.reverse()

#눈금선
plt.style.use('ggplot')

#해상도
plt.rcParams["figure.dpi"] = 300

#글씨 깨짐 방지
plt.rc('font', family = 'Malgun Gothic')

#왼쪽 '-'' 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

#세로 막대 그래프
plt.barh(range(101), male, label = '남성', color = 'skyblue')
plt.barh(range(101), female, label = '여성', color = 'lightpink')

#범례
plt.legend()
plt.title(header)
plt.show()
```
