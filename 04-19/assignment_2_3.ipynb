```python
import matplotlib.pyplot as plt
import csv

f = open('seoul.csv')
data = csv.reader(f)
next(data)

days = []
for i in range(1, 32):
    days.append([])

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08':
            idx = int(row[0].split('-')[2])
            days[idx-1].append(float(row[-1]))

plt.rc('font', family = 'Malgun Gothic')
plt.title('8월의 기온분포')
plt.boxplot(days)
plt.show()
```
