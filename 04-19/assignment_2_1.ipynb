```python
import matplotlib.pyplot as plt
import csv

f = open('seoul.csv')
data = csv.reader(f)
next(data)

high = []
low = []

for row in data:
    if row[-1] != '':
        if int(row[0].split('-')[0]) >= 1983:
            if int(row[0].split('-')[1]) == 2 and int(row[0].split('-')[2]) == 14:
                high.append(row[-1])
                low.append(row[-2])


plt.rc('font', family = 'Malgun Gothic')
plt.title('내 생일의 기온 변화 그래프')
plt.plot(high, 'lightpink', label = '최고기온')
plt.plot(low, 'skyblue', label = '최저기온')
plt.legend()
plt.show()
```
